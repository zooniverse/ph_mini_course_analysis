```python
from  pymongo import MongoClient
from pylab import *
%matplotlib inline
```

```python
client = MongoClient()
db = client.ouroboros
```

```python
planet_hunter_proj_id = "5333063b3ae740228a000001"
planet_hunter_users = db.planet_hunter_users
```

#Basic stats 
These are some basic stats about how many users opted in and out of the minicoruse

```python
total_users = planet_hunter_users.count()
no_opt_in  = planet_hunter_users.find({"preferences.planet_hunter.course" : "yes"}).count()
no_opt_out = planet_hunter_users.find({"preferences.planet_hunter.course" : "no"}).count()

print str(no_opt_in) +", "+ str(no_opt_in*100.0/total_users) + "% users opted in"
print str(no_opt_out) +", "+ str(no_opt_out*100.0/total_users) +"% users opted out"
```

#Course progress 

These plots show for the users who took the course how much they participated. So for example about 65% of users who opted in to the course viewed only 1 slide with only roughtly 5% completing the coruse

```python
course_progress  = [int(ph["preferences"]["planet_hunter"]["curr_course_id"]) for ph in planet_hunter_users.find({"preferences.planet_hunter.curr_course_id" : {"$exists" : True} })]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(24,8))

axes[0].set_title("Minicourse use")
axes[0].set_xlabel("No of course slides viewed")
axes[0].set_ylabel("No of users")
axes[0].hist(course_progress,bins=25)


axes[1].set_title("Minicourse use")
axes[1].set_xlabel("No of course slides viewed")
axes[1].set_ylabel("% of  users")
axes[1].hist(course_progress, normed=True, bins=25)   
```

```python


fig = plt.figure()
ax = plt.gca()
ax.set_yscale('log')
ax.set_title("Course participation vs classification count")
ax.set_xlabel("No of course slides done")
ax.set_ylabel("Classification count")

classification_counts   = [int(ph["projects"][planet_hunter_proj_id].get("classification_count", 0)
                               ) for ph in planet_hunter_users.find({"preferences.planet_hunter.curr_course_id" : {"$exists" : True} })]
ax.plot(course_progress, classification_counts, 'o')

mean_fig = plt.figure()
mean_fig_ax = mean_fig.gca()
together = zip(course_progress, classification_counts) 
means = [median([c_count for course_no,c_count in together if course_no == i]) for i in [0,5,10,15,25] ]
stds  = [std([c_count for course_no,c_count in together if course_no == i]) for i in [0,5,10,15,25] ]

mean_fig_ax.errorbar([0,5,10,15,25], means, fmt="o", yerr=stds)
mean_fig_ax.set_yscale('log')
```

```python


hist_fig = plt.figure()
hist_fig_ax = hist_fig.gca()




for i in [0,5,10,15,25]:
    hist_fig_ax.hist( [c_count for course_no,c_count in together if course_no == i], normed=True, range=[0,4000], bins=20, log=True, label=str(i)+" course views")
hist_fig_ax.legend()

```
